```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display full output in same cell
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

%matplotlib inline

# 🔹 Load Dataset (Space separated fix)
df = pd.read_csv("../datasets/house Prediction Data Set.csv",
                 sep=r"\s+",
                 header=None)

# 🔹 Assign Column Names (Boston Housing Standard)
df.columns = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE",
              "DIS","RAD","TAX","PTRATIO","B","LSTAT","PRICE"]

print("===== FIRST 5 ROWS =====")
print(df.head())

print("\n===== SHAPE =====")
print(df.shape)

print("\n===== DATASET INFO =====")
print(df.info())

print("\n===== SUMMARY STATISTICS =====")

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(df.describe().T)


print("\n===== MISSING VALUES =====")
print(df.isnull().sum())

# 🔥 Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 🔥 Histogram of all features
df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()

# 🔥 Scatter Plot 1
plt.figure(figsize=(6,4))
sns.scatterplot(x="RM", y="PRICE", data=df)
plt.title("Rooms vs Price")
plt.show()

# 🔥 Scatter Plot 2
plt.figure(figsize=(6,4))
sns.scatterplot(x="LSTAT", y="PRICE", data=df)
plt.title("LSTAT vs Price")
plt.show()

print("\n✅ TASK 3 COMPLETED SUCCESSFULLY")
```
